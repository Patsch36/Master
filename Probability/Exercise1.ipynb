```python
from scipy.stats import binom

def pdf_num_heads(num_dice):
    p = 1/6  # Wahrscheinlichkeit, einen bestimmten Wert auf dem Würfel zu erhalten
    n = num_dice  # Anzahl der Münzwürfe, die von den Würfelwürfen abhängt
    dist = binom(n, p)
    return dist.pmf(range(n+1))

# Beispielaufruf:
num_dice = 6  # Anzahl der Münzwürfe
probabilities = pdf_num_heads(num_dice)
for y, prob in enumerate(probabilities):
    print(f'Anzahl Köpfe: {y}, Wahrscheinlichkeit: {prob}')
```

```python
import matplotlib.pyplot as plt

plt.bar(range(len(probabilities)), probabilities)
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.title('Probability Distribution of Number of Heads')
plt.show()
```
